```python
pokemon = pd.read_csv('../../../datasets/Pokemon.csv', index_col=0)
pokemon.head()
```

```python
pokemon.plot.scatter(x='Attack',y='Defense')
```

```python
pokemon.plot.hexbin(x='Attack',y='Defense', gridsize=30)
```

```python
pokemon_group = pokemon.groupby(['Legendary','Generation']).mean()[['Attack', 'Defense']]
pokemon_group.head()
```

```python
pokemon_group.plot.bar(stacked=True)
```

```python
pokemon_group.describe()
```

```python
pokemon_time = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokemon_time.describe()
```

```python
pokemon_time.plot.line()
```
