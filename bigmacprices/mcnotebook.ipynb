```python
df = df.drop(['iso_a3', 'currency_code', 'local_price', 'yen_valuation','dollar_ex','dollar_ppp','euro_adj_valuation','yuan_valuation','sterling_valuation', 'GDP_dollar', 'dollar_valuation', 'yuan_adj_valuation','yen_adj_valuation','sterling_adj_valuation' ,'dollar_adj_valuation', 'euro_valuation'], axis=1)
```

```python
df.head(5)
```

```python
import os
```

```python
state_geo = os.path.join('../../folium/examples/data/', 'world-countries.json')
```

```python
state_geo
```

```python
m = folium.Map(location=[-23, -46], zoom_start=3)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=df,
 columns=['Country', 'dollar_price'],
 key_on='feature.properties.name',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Big Mac Price'
)
folium.LayerControl().add_to(m)
 
# Save to html
m.save('folium_chloropleth_bigmac.html')
```
