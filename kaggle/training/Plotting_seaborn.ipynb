```python
import pandas as pd
wines = pd.read_csv('../../../datasets/winemag-data_first150k.csv')
import seaborn as sns
```

```python
sns.countplot(wines['points'])
```

```python
sns.kdeplot(wines.query('price < 200').price)
```

```python
wines[wines['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

```python
wines_sample = wines.sample(5000)
sns.kdeplot(wines_sample.query('price < 200').price.dropna(), wines_sample.query('price < 200').points.dropna(), 
            shade=True, gridsize=30)
```

```python
sns.distplot(wines['points'], bins=10, kde=False)
```

```python
sns.jointplot(x='price', y='points', data=wines[wines['price'] < 100])
```

```python
sns.jointplot(x='price', y='points', data=wines[wines['price'] < 100], kind='hex', gridsize=20)
```

```python
df = wines[wines.variety.isin(wines.variety.value_counts().head(5).index)]

sns.boxplot(x='variety',y='points',data=df, width=0.5)
```

```python
sns.violinplot(
    x='variety',
    y='points',
    data=wines[wines.variety.isin(wines.variety.value_counts()[:5].index)]
)
```
